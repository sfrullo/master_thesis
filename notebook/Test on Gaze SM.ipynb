```python
data = np.load("../export/data/gaze_sm_1811260944.npz")["gaze_sm"]
```

```python
from sklearn import preprocessing

import gc
gc.collect()

normalized_data = data.copy()
normalized_data[:,0] = preprocessing.minmax_scale(normalized_data[:,0])
print normalized_data.shape
plt.plot(normalized_data[:,200/2,320/2])
plt.show()

#print normalized_data[:,0].shape
#print np.where(normalized_data[:,0] == 1)
#print normalized_data[:,13,92]
```

```python
frame_scale_factor=4

dpi = 100
bins = np.array([1280, 800])/frame_scale_factor

ncols=2
nrows=1

# determine the figure size in inches
figsize = (bins[0]*2*ncols/dpi, bins[1]*2*nrows/dpi)

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, subplot_kw=dict(xlim=[0,bins[0]], ylim=[0,bins[1]]), figsize=figsize, dpi=dpi)

f = 17

axes[0].imshow(data[f])
axes[1].imshow(normalized_data[f])
plt.show()
```
