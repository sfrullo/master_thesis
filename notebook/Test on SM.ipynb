```python
%matplotlib inline

import os
import glob

import numpy as np

import matplotlib as mpl
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] =(12,9)
```

```python
data = np.load("export/data/gaze_sm_1811111158.npz")["gaze_sm"]
```

```python
normalized_data = data.copy()
normalized_data = (normalized_data - normalized_data.mean())/normalized_data.max()
```

```python
frame_scale_factor=4

dpi = 100
bins = np.array([1280, 800])/frame_scale_factor

ncols=2
nrows=1

# determine the figure size in inches
figsize = (bins[0]*2*ncols/dpi, bins[1]*2*nrows/dpi)

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, subplot_kw=dict(xlim=[0,bins[0]], ylim=[0,bins[1]]), figsize=figsize, dpi=dpi)

f = 50

axes[0].imshow(data[f])
axes[1].imshow(normalized_data[f])
plt.show()
```
