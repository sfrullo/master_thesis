```python
import glob
import xmltodict
import numpy as np
import pandas as pd
```

```python
list_of_session = []
```

```python
for file in glob.glob("../Dataset/Mahnob/data/Sessions/**/*.xml"):
    try:
        d = xmltodict.parse(open(file).read())["session"]
        list_of_session.append({
            "sessionId" : d["@sessionId"],
            "date" : d["@date"],
            "time" : d["@time"],
            "experimentType" : d["@experimentType"],
            "isStim" : d["@isStim"],
            "feltEmo" : int(d["@feltEmo"]),
            "feltArsl" : int(d["@feltArsl"]),
            "feltVlnc" : int(d["@feltVlnc"]),
            "feltCtrl" : int(d["@feltCtrl"]),
            "feltPred" : int(d["@feltPred"]),
            "cutNr" : d["@cutNr"],
            "cutLenSec" : d["@cutLenSec"],
            "vidRate" : d["@vidRate"],
            "audRate" : d["@audRate"],
            "audBeginSmp" : d["@audBeginSmp"],
            "audEndSmp" : d["@audEndSmp"],
            "vidBeginSmp" : d["@vidBeginSmp"],
            "vidEndSmp" : d["@vidEndSmp"],
            "mediaFile" : d["@mediaFile"],
        })
    except Exception, e:
        print "skipping {}, error:".format(file, e.message)
```

```python
files = map(np.load, glob.glob("../export/data/feature_data_*_with_pupil.npz"))

sessions_to_skip = {
    "30"   : [800, 1178, 1712, 2628, 3646],
    "53"   : [426, 814, 3028, 3538],
    "69"   : [1726, 2866, 3134, 3670],
    "90"   : [664, 932, 1714, 3012],
    "111"  : [32, 1210, 2354, 3158]
}


valid_files = [ 
    f["session_info"].tolist()
    for f in files 
    if f["session_info"].tolist().get("mediaFile").split(".")[0] in sessions_to_skip.keys()
]
```

```python
a = pd.DataFrame(valid_files)
a['mediaFile'] = a['mediaFile'].map(lambda x: int(x.strip('.avi')))
a
```

```python
t = a[["sessionId", "mediaFile"]].groupby("mediaFile").count()
print t.sum().to_latex()
```

```python
b = a[["mediaFile", "feltEmo", "feltArsl", "feltVlnc"]].applymap(int)
print b.groupby('mediaFile').agg({'feltEmo':['mean', 'std'], 
                         'feltArsl':['mean', 'std'],
                         'feltVlnc':['mean', 'std']}).round(2).to_latex()
```
