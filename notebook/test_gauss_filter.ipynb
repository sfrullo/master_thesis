```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.pyplot import cm
from matplotlib.colors import ListedColormap

from scipy.ndimage import gaussian_filter
```

```python
base = np.zeros([200, 320], dtype=np.float)
```

```python
plt.imshow(base, cmap=cm.gist_gray)
```

```python
# IMAGE 1
```

```python
im1 = base.copy()
im1[175,50] = 1
plt.imshow(im1, cmap=cm.gray)
```

```python
im1g = im1.copy()
im1g = gaussian_filter(im1g, sigma=25)
plt.imshow(im1g, cmap=cm.gray)
```

```python
# IMAGE 2
```

```python
im2 = base.copy()
im2[150,200] = 1
plt.imshow(im2, cmap=cm.gray)
```

```python
im2g = im2.copy()
im2g = gaussian_filter(im2g, sigma=25)
plt.imshow(im2g, cmap=cm.gray)
```

```python
# FUSION
```

```python
base_im1 = np.add(base,im1)
base_im1_im2 = np.add(base_im1,im2)
plt.imshow(base_im1_im2, cmap=cm.gray)
```

```python
base_im1 = np.add(base,im1)
base_im1_im2 = np.add(base_im1,im2)
base_im1_im2_g = gaussian_filter(base_im1_im2, sigma=25)
plt.imshow(base_im1_im2_g, cmap=cm.gray)
```

```python
base_im1g = np.add(base,im1g)
base_im1g_im2g = np.add(base_im1g,im2g)
plt.imshow(base_im1g_im2g, cmap=cm.gray)
```

```python
from math import atan2, degrees

def compute_visual_angle(distance):
    h = 51.9 # Monitor height in cm
    d = distance # Distance between monitor and participant in cm
    r = 1280 # Vertical resolution of the monitor
    # Calculate the number of degrees that correspond to a single pixel. This will
    # generally be a very small value, something like 0.03.
    deg_per_px = degrees(atan2(.5*h, d)) / (.5*r)
    return deg_per_px
```

```python
compute_visual_angle(693.117900)
```
